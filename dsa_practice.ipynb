```python
# 1. Pair Sum Count (Unique Pairs)
# Problem:
# Given an array of integers and a target value, count the number of unique pairs (i, j) such that arr[i]+arr[j] = target.
# Constraints:
# Same pair should not be counted twice
# Order does not matter

# Example:
# Input: arr = [1, 4, 5, 2, 6, 7, 8, 8], target = 10
# Output: 2   # (2,8), (4,6)


def find_pairs(arr, target):
    combinations = []
    temp_dict = {}
    for i, num in enumerate(list(set(arr))):
        required = target - num
        if required in temp_dict:
            combinations.append([num, required])
        temp_dict[num] = i 

    return len(combinations)

arr = [1, 4, 5, 2, 6, 7, 8, 8] 
target = 10
print(find_pairs(arr, target))

#or [more optimized version]
# Option 1: Using a set (Recommended)

def find_pair(arr, target):
    seen = set()
    count = 0
    for num in set(arr):
        if target-num in seen:
            count+=1
        seen.add(num)
    return count

print(find_pair(arr, target))
        
```

```python
# 7. Top K Frequent Elements
# Problem:
# Given an integer array, return the K most frequent elements.
# Example:
# Input: nums = [1,1,1,2,2,3], k = 2
# Output: [1,2]

from collections import Counter
def top_k(arr, k):
    # DEGUB
    print(Counter(arr).most_common(k))  #[(1, 3), (2, 2)] so extract required num, we dont need freq
    return [num for num, freq in Counter(arr).most_common(k)]

nums = [1,1,1,2,2,3] 
k = 2
print(top_k(nums,k))
```

```python
# 9. Check If Two Strings Are Isomorphic
# Problem:
# Two strings are isomorphic if characters in one string can be replaced to get the other string.

# Example:
# Input: s = "egg", t = "add"
# Output: True

from collections import Counter
def is_isomorphic(s,t):
    freq_s = [freq for char, freq in dict(Counter(s)).items()]
    freq_t = [freq for char, freq in dict(Counter(t)).items()]

    print(freq_s)
    print(freq_t)

    

    return freq_s == freq_t
    
s = "egg"
t = "add"
print(is_isomorphic(s,t))
```
