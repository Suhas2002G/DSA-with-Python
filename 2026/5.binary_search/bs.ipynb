```python
# Given a sorted array of integers arr and an integer target, write a function to find the starting index and ending index of the given target value.
# If the target is not present in the array, return [-1, -1]
# Input:
# arr = [1, 1, 2, 2, 2, 2, 3, 4, 5]
# target = 2
# Output:
# [2, 5]

# input:
# arr = [1, 3, 5, 7, 9]
# target = 4
# output:
# [-1, -1]

def find_first(arr, target):
    low, high = 0, len(arr)-1
    first = -1
    
    while low <= high:
        mid = (low+high)//2
        if arr[mid] == target:
            first = mid 
            high = mid-1
        elif arr[mid] < target:
            low = mid+1
        else:
            high = mid-1
    return first 

def find_last(arr, target):
    low, high = 0, len(arr)-1
    last = -1
    
    while low <= high:
        mid = (low+high)//2
        if arr[mid] == target:
            last = mid 
            low = mid+1
        elif arr[mid] < target:
            low = mid+1
        else:
            high = mid-1
    return last 

def solution(arr, target):
    first = find_first(arr, target)
    if first == -1:
        return [-1, -1]
    last = find_last(arr,target)
    return [first, last]
            
            
arr = [1,1,2,2,2,2,3,4,5]
print(solution(arr,2))
```
