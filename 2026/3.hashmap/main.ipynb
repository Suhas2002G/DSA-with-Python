```python
# Q1. Find frequency of each element in the array
# Input: arr = [1,2,2,3,1,1]
# Output: {1:3, 2:2, 3:1}
# Time Complexity: O(n)
# Space Complexity: O(n)

def find_frequency(arr: list)->dict:
    freq = {}
    for i in arr:
        freq[i] = freq.get(i,0) + 1
    return freq 

arr = [1,2,2,3,1,1]
print(find_frequency(arr))

#OR

from collections import Counter
def find_frequency(arr: list)->dict:
    return dict(Counter(arr)) 

arr = [1,2,2,3,1,1]
print(find_frequency(arr))
```

```python
# Q2. Check if array contains duplicate
# Input: arr = [1,2,3,4,1]
# Output: True

def is_duplicate(arr: list)->bool:
    freq = {}
    for i in arr:
        freq[i] = freq.get(i,0) + 1
    
    for k,v in freq.items():
        if v > 1:
            return True
    return False

arr = [1,2,3,4,1]
print(is_duplicate(arr))

#OR -> BEST APPROACH

def is_duplicate(arr: list)->bool:
    seen = set()
    for num in arr:
        if num in seen:
            return True
        seen.add(num)
    return False

arr = [1,2,3,4,1]
print(is_duplicate(arr))
```

```python
# Q3. Find the first repeating element
# Input: arr = [10,5,3,4,3,5,6]
# Output: 3

def find_first_repeated_ele(arr):
    seen = set()
    for num in arr:
        if num in seen:
            return num
        seen.add(num)
    return -1 # if repeted element not found return -1

arr = [10,5,3,4,3,5,6]
print(find_first_repeated_ele(arr))
```

```python
# Q4. Find the first non-repeating element
# Input: arr = [9,4,9,6,7,4]
# Output: 6

def first_unique(arr):
    freq = dict(Counter(arr))
    for k,v in freq.items():
        if v == 1:
            return k 
        

arr = [9,4,9,6,7,4]
print(first_unique(arr))
```

```python
# Q5. Two Sum
# Input: arr = [2,7,11,15], target = 9
# Output: [0,1]

def two_sum(arr, target):
    result = {}
    for i,num in enumerate(arr):
        required = target - num 
        if required in result:
            return [result[required], i]
        result[num] = i 

arr = [2,7,11,15]
target = 9
print(two_sum(arr, target))
```

```python
# Q6. Count number of pairs {i,j} with sum = K
# Input: arr = [1,5,7,-1,5], k = 6
# Output: 3

from itertools import combinations

def count_pairs(arr, k):
    count = 0
    pairs = list(combinations(arr,2))

    for pair in pairs:
        if pair[0] + pair[1] == k:
            count +=1
    return count


arr = [1,5,7,-1,5] 
k = 6
print(count_pairs(arr,k))
```

```python
# Q8. Check if two arrays are equal
# Input:
# arr1 = [1,2,2,3]
# arr2 = [2,1,3,2]
# Output: True
# NOTE: if freq of all elemnts from both array is same, then they are same

def is_equal(arr1, arr2):
    return Counter(arr1) == Counter(arr2)

arr1 = [1,2,2,3]
arr2 = [2,1,3,2]
print(is_equal(arr1, arr2))
```

```python
# Find element with maximum frequency
# Input: arr = [1,3,2,1,4,1]
# Output: 1
def max_frequency(arr):
    freq = {}
    for num in arr:
        freq[num] = freq.get(num, 0) + 1
    return max(freq, key=freq.get)

arr = [1,3,2,1,4,1]
print(max_frequency(arr))

#OR

def max_frequency(arr):
    return Counter(arr).most_common(1)[0][0]

arr = [1,3,2,1,4,1]
print(max_frequency(arr))


#OR

# def max_frequency(arr):
#     freq = Counter(arr)
#     curr = 0
#     max = 0
#     print(freq)
#     for k,v in freq.items():
#         if v > curr:
#             max = k
#             curr = v
        
#     return max

# arr = [1,3,2,1,4,1]
# print(max_frequency(arr))
```

```python
# Q10. Find missing number
# Input: arr = [1,2,3,5]
# Output: 4

def missing_number(arr):
    n = len(arr)
    actual_sum = sum(arr)
    # print(actual_sum)

    required_sum = int(n+1*(n+2) / 2)
    # print(required_sum)

    return actual_sum - required_sum

arr = [1,2,3,5]
print(missing_number(arr))
```
