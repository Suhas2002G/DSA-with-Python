```python
l1 = [1,2,3,4]
l2 = [5,4,9,2]
d = dict(zip(l1,l2))
d
```

```python
# map method
arr = [1,2,3,4,5]
sq = list(map(lambda x: x*x,arr))
sq
```

```python
# filter 
even = list(filter(lambda x : x %2 == 0, arr))
even
```

```python
# reduce 
from functools import reduce 

total = reduce(lambda x,y: x+y, arr)
total 
```

```python
from itertools import combinations

pairs = list(combinations(arr,2))
pairs
```

```python
filter_pair = list(filter(lambda x:x[0]+x[1]==6, pairs))
filter_pair
```
