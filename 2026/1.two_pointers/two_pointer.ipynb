```python
# Problem: Pair Sum (Classic)
# Question: Given a sorted array and a target sum, find if any two numbers add up to target.

def two_sum(arr, target):
    start, end = 0, len(arr)-1

    while start < end:
        total = arr[start] + arr[end]
        if total == target:
            return [start,end]
        elif total > target:
            end -= 1
        elif total < target:
            start +=1

    return [-1,-1]


arr = [1, 2, 3, 4, 7]
target = 7
print(two_sum(arr, target))
```

```python
# 1. Reverse an Array
# Problem Statement: Given an array of integers, reverse the array using the two-pointer technique.
# Input: arr = [1, 2, 3, 4, 5]
# Output: [5, 4, 3, 2, 1]

def reverse_array(arr):
    start, end = 0, len(arr)-1

    while start<end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1

    return arr


arr = [1, 2, 3, 4, 5]
print(reverse_array(arr))
```

```python
# 2. Check Palindrome String
# Problem Statement: Check whether a given string is a palindrome using two pointers.
# Input: s = "madam"
# Output: True

def is_pallindrome(str):
    start, end = 0, len(str)-1

    while start < end:
        if str[start] != str[end]:
            return False
        start += 1
        end -= 1
    return True

print(is_pallindrome('madam'))
```

```python
# 4. Remove Duplicates from Sorted Array
# Problem Statement: Remove duplicates from a sorted array in-place and return new length.
# Input: arr = [1, 1, 2, 2, 3]
# Output: 3

def remove_duplicate_sorted_array(arr):
    if not arr:
        return 0
    
    k = 0 # represents the last index of a unique element
    
    for i in range(1, len(arr)):
        if arr[k] != arr[i]:
            k+=1
            arr[k] = arr[i]  # overwrite duplicate

    return k+1, arr[:k+1]

arr=[1, 1, 2, 2, 3, 4, 5, 5]
print(remove_duplicate_sorted_array(arr))
```

```python
# 5. Move Zeros to End
# Problem Statement: Move all zeros to the end while maintaining order of non-zero elements.

# Input: arr = [0, 1, 0, 3, 12]
# Output: [1, 3, 12, 0, 0]

def move_zeros(arr):
    k = 0  # k is our 'Collector'
    
    for i in range(len(arr)):
        if arr[i] != 0:
            arr[k], arr[i] = arr[i], arr[k]
            
            # Move (k) forward to the next empty-looking spot
            k += 1
            
    return arr

arr = [0, 1, 0, 3, 12, 0, 13]
print(move_zeros(arr))
```

```python
# 6. Squares of Sorted Array
# Problem Statement: Return squares of each number sorted in non-decreasing order.
# Input: arr = [-4, -1, 0, 3, 10]
# Output: [0, 1, 9, 16, 100]

def sq_sorted_arr(arr):
    n = len(arr)
    result = [0] * n 
    left, right = 0, n-1
    pos = n-1

# Using <= ensures we don't forget the very last number in the middle!
    while left <= right:
        left_sq = arr[left]**2
        right_sq = arr[right]**2
        
        if left_sq > right_sq:
            result[pos] = left_sq
            left += 1
        else:
            # This covers if right is bigger OR if they are equal
            result[pos] = right_sq
            right -= 1

        pos -= 1

    return result 

arr = [-4, -1, 0, 3, 10]
print(sq_sorted_arr(arr)) 
# Output: [0, 1, 9, 16, 100]
```

```python
# 7. Valid Palindrome (Ignore Non-Alphanumeric)
# Input
# s = "A man, a plan, a canal: Panama"
# Output
# True

def valid_pallindrome(str):
    start = 0
    end = len(str)-1

    str = str.lower()   # NOTE: check given string upper case and lower case, A != a

    while start < end:
        if not str[start].isalnum():
            start +=1
            continue
        if not str[end].isalnum():
            end -=1
            continue
        
        if str[start] != str[end]:
            return False 
        
        start +=1 
        end -=1

    return True

str = "A man, a plan, a canal: Panama"
print(valid_pallindrome(str))
```
