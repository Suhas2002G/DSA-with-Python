```python
# Two Sum
# Given an array of integers nums and an integer target, return indices of two numbers that add up to target.
# Input: nums = [2,7,11,15], target = 9 → Output: [0,1]

def two_sum(input: list[int], target: int):
    result = {}
    for i, num in enumerate(input):
        req = target - num
        if req in result:
            return [result[req], i]
        result[num] = i


input = [2,7,11,15]
target = 9
print(two_sum(input, target))
```

```python
# Contains Duplicate
# Return True if any value appears at least twice.

def is_duplicate(l1):
    return len(l1) > len(set(l1))


l1 = [1,2,3,4,5,5]
print(is_duplicate(l1))
```

```python

def binary_search(nums: list[int], target: int)-> int:
    '''
    To find the index of target integer using Binary search algorithm
    NOTE: BSA is done only in sorted list
    params:
        nums : list of integeers
        target: integer to find its index

    returns:
        result: int -> the index of the target integer 
    '''

    start = 0
    end = len(nums)-1

    while start <= end:
        mid = (start+end)//2
        if nums[mid] == target:
            return mid

        elif nums[mid] < target:
            start = mid + 1

        else:
            end = mid - 1

    return -1
```

```python
nums = [1,2,23,43,234,4234]
target = 2
print(binary_search(nums, target))
```
