```python
def binary_search(nums, target):
    start = 0
    end = len(nums) - 1

    while start <= end:
        mid = (start + end) // 2

        if nums[mid] == target:
            return mid
        
        elif nums[mid] < target:      # go right
            start = mid + 1
        
        else:                         # nums[mid] > target â†’ go left
            end = mid - 1

    return -1

nums = [1,3,4,6,7,8,9,12]
target = 8
print(binary_search(nums, target))
```
