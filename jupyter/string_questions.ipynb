```python
# Q1. Write a function that counts how many times a specific character appears in a string

# time complexity : O(n)
# approach 1
def character_counter(ch: str, s: str)->int:
    count = 0
    for i in s:
        if i == ch:
            count += 1 
    return count

# approach 2: Using Built-in count()
def character_counter(ch: str, s: str)->int:
    return s.count(ch)

print(character_counter('s','suhas'))
```

```python
# Q2. Create a function that reverses a string without using built-in reverse methods

def string_reverser(s: str)->str:
    reverse_str = ''
    for i in s:
        reverse_str = i + reverse_str 
    return reverse_str 
    
print(string_reverser('suhas')) 

# Time Complexity: O(nÂ²)
# Space Complexity: O(n)
# Why your approach is suboptimal:
# Your solution repeatedly creates new strings using i + reverse_str, and 
# because strings are immutable in Python, each concatenation takes O(n) time. 
# This results in quadratic time complexity, making the approach inefficient for large input sizes.

#--------------------------------------------------------------------------------------------------

def string_reverser2(s: str)->str:
    reverse = [] 
    for i in s:
        reverse.insert(0,i)
    # print(reverse) # ['s', 'a', 'h', 'u', 's']
    return ''.join(reverse)
    
print(string_reverser2('suhasgondukupi'))

# Time Complexity: O(n)
# Space Complexity: O(n)
# Why this approach is better:
# This method iterates backward through the string and uses list append operations, 
# which are O(1). Joining the list at the end is O(n), resulting in a linear-time solution that is more 
# efficient and scalable than repeated string concatenation.
    
```

```python
# Q3. Write a program that checks if a string is a palindrome

#------------ 2 pointer approach --------------
def pallindrome_checker(s: str)->bool:
    start = 0
    end = len(s)-1 

    while start < end:
        if s[start] != s[end]:
            return False 
        start += 1 
        end -= 1
    return True

print(pallindrome_checker('suhus'))

# Time Complexity: O(n), Space Complexity: O(1)
# Why your approach is good but can be improved:
# Your solution efficiently uses a two-pointer technique and correctly achieves linear time with constant space. 
# However, it manually manages pointers, making the code longer and less readable compared to more concise approaches.


#------------ slicing --------------
def pallindrome_checker(s: str)->bool:
    return s == s[::-1]

print(pallindrome_checker('suhus'))

# Time Complexity: O(n), Space Complexity: O(n)
# Why this approach is better:
# This solution avoids manual pointer handling and leverages slicing to reverse the string in a single, concise expression. 
# It improves readability and maintains the same linear time complexity while being more maintainable.
    
```

```python
# Q6. Write a function that finds the longest word in a string

def longest_word(sentence: str)-> str:
    longest = ''
    words = sentence.split()
    for i in words:
        if len(i) > len(longest):
            longest = i 
    return longest


sentence = 'suhas is full stack developer'
print(longest_word(sentence))

# Time Complexity: O(n)
# Space Complexity: O(n)
```
