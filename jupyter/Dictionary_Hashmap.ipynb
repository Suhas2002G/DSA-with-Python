##### EASY (10 Questions)

```python
# Q1. Count frequency of each character in a string
# Input: "pythonprog", Output: {'p':2, 'y':1, ...}

from collections import Counter

def frequency_counter(s: str)->dict:
    freq = dict(Counter(s.lower()))
    return freq

print(frequency_counter('pythonprog'))
```

```python
# Q2. Find the first non-repeating character in a string
# Input: "swiss" → Output: "w"

from collections import Counter

def non_repeat_ele(s: str)->str:
    freq = dict(Counter(s.lower()))

    for k in freq:
        if freq[k] == 1:
            return k    
    return None
    


s = 'Swiss'
print(non_repeat_ele(s))
    
```

```python
# Q3. Check if two strings are anagrams
# Input: "listen", "silent" → Output: True

from collections import Counter

# def check_anagram(s1: str, s2: str)->bool:
#     if len(s1) != len(s2):
#         return False

#     d1 = dict(Counter(s1.lower()))
#     d2 = dict(Counter(s2.lower()))

#     if d1 == d2:
#         return True
#     return False

    
#OR [more shorter version]
def check_anagram(s1: str, s2: str)->bool:
    return Counter(s1.lower()) == Counter(s2.lower())


s1 = 'listen'
s2 = 'silent'
print(check_anagram(s1,s2))
```

```python
# Q4. Count frequency of each word in a sentence

from collections import Counter

def count_word_freq(sentence: str)-> dict:
    return dict(Counter(sentence.lower().split()))

sentence = 'python is easy and easy but java is hard'
print(count_word_freq(sentence))
```

```python
# Q5. Return the key with max frequency

str = 'usuhas'
d = dict(Counter(str.lower()))

max_freq = max(d.values())                          # highest count
print(max_freq)
result = [k for k, v in d.items() if v == max_freq] # all keys with that freq

print(result)
```

```python
# Q6. Check if two lists contain same elements with same frequency

from collections import Counter

l1 = ['cat', 'bat', 'mat', 'cat']
l2 = ['cat', 'at', 'mat', 'cat']

print(Counter(l1) == Counter(l2))
```

```python
# Q7. Convert 2 lists into a hashmap
# Keys: [‘name’, ‘age’] # Values: [‘suhas’, 21]

keys = ['name', 'age']
values = ['suhas', '23']

result = dict(zip(keys, values))
print(result)
```

```python
#Q1. Two Sum using HashMap
# Input: [2,7,11,15], target=9
# Output: indices [0,1]

input = [2,7,11,15]
target = 9

def two_sum(nums, target):
    d = {}
    for i, num in enumerate(nums):
        req = target - num        # find complement
        if req in d:              # check if complement exists
            return [d[req], i]    # return indices
        d[num] = i                # store current number with index

print(two_sum(input,target))
            
```

```python
#Q2. Find duplicates in an array using HashMap

from collections import Counter

input = [2,7,11,15,15,2,1]

def find_duplicates(nums):
    freq = dict(Counter(nums))
    # print(freq)  # {2: 2, 7: 1, 11: 1, 15: 2, 1: 1}

    dup = []
    for i in freq:
        if freq[i] > 1:
            dup.append(i)
    return dup

print(find_duplicates(input))
```
