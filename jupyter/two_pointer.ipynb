#### Finding a Perfect Gift Pair
> You run a gift shop and customers want to buy two items whose total price matches their gift budget exactly.
> You have a sorted price list: [5, 10, 15, 20, 25, 30]
> Budget = 35
> Problem: Find any pair of prices whose sum equals exactly 35.

```python
def gift_finder(price_list: list, budget: int)-> list[int]:
    start = 0
    end = len(price_list) - 1 

    while start < end:
        sum = price_list[start] + price_list[end]
        if sum == budget:
            return [start, end]
        elif sum < budget:
            start += 1
        elif sum > budget:
            end -= 1
        else:
            return []

# Test code:
price_list = [5, 10, 15, 20, 25, 30]
budget = 45

print(gift_finder(price_list, budget))
    
```
