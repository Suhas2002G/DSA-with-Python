### List 

```python
# Q1. Write a program to find the sum of all elements in a list.

def sum_of_elements(lst: list)->int:
    return sum(lst)

lst=[1,2,4,5,3]
print(sum_of_elements(lst))
```

```python
# Q2. Remove duplicates from a list while keeping order.

lst=[1,2,4,4,5,3,1,2,8,9]

unique = []
for i in lst:
    if i not in unique:
        unique.append(i)
print(unique)
```

```python
# Q3. Find the second largest element in a list.

def find_second_largest_ele(lst):
    # create a copy of original list, since sort() modify existing list instead of creating new list
    list_by_order = lst
    list_by_order.sort()
    return lst[-2]


lst=[1,2,4,56,4,5,3,1,2,8,9]
print(find_second_largest_ele(lst))
        
```

#### Dict

```python
# Q1. Create a dict from two lists: keys = ['a','b'], values=[1,2].

def create_dict_from_list(l1,l2):
    # return zip(l1,l2)   # <zip object at 0x000001E5534A2640>
    return dict(zip(l1,l2))

keys = ['a','b'] 
values=[1,2]
print(create_dict_from_list(keys, values))
```

```python
# Q2. Merge two dictionaries.

d1={'a':1, 'b':2}
d2={'c':3, 'd':4}

# op1: Using the Dictionary Union Operator (|)
merged_dict = d1 | d2

# op2: Unpack both dictionaries into a new one
merged_dict = {**d1, **d2}

# op3: use update() without modifying an existing dictionary, you can first make a copy of the first dictionary
merged_dict = d1.copy()
merged_dict.update(d2)

print(merged_dict)
```
