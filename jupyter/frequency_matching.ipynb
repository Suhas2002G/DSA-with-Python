```python
# Q. 2. Find the most frequent character

str = 'pythonprogp'

from collections import Counter

freq = dict(Counter(str))
print(freq)

max_freq = max(freq, key=freq.get)
max_freq
```

```python
min_freq = min(freq.values(), key=freq.get)
min_freq
```
