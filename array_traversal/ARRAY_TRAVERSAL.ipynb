```python
# Find Maximum Element in an Array
# Problem Statement: Given an array of integers, find and return the maximum element present in the array.
# Do not use built-in functions like max().
# Input: [3, 5, 1, 9, 2]
# Output: 9

def find_max(arr):
    maximum = arr[0]
    for i in arr:
        if i > maximum:
            maximum = i
    return maximum

print(find_max([3, 5, 1, 9, 2]))
```

```python
# Q3. Sum of All Elements in an Array
# --> Problem Statement: Calculate the sum of all elements present in an integer array.
# --> Example: Input: [1, 2, 3, 4], Output: 10

def find_sum(arr):
    sum = 0
    for i in arr:
        sum += i
    return sum

print(find_sum([1, 2, 3, 4]))
```

```python
# Count Even Numbers in an Array
# Problem Statement: Count how many even numbers are present in the given array.
# Example: Input: [1, 2, 4, 7, 8], Output: 3

def even_counter(arr):
    count = 0
    for i in arr:
        if i % 2 == 0:
            count += 1
    return count
    
arr = [1, 2, 4, 7, 8]
print(even_counter(arr))
```

```python
# Reverse an Array (In-Place)
# Problem Statement: Reverse the array without using extra space.
# Example: Input: [1, 2, 3, 4], Output: [4, 3, 2, 1]
# i used 2-pointer approach

def reverse_array(arr):
    start = 0
    end = len(arr)-1
    while start < end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1
    return arr

print(reverse_array([1, 2, 3, 4, 5]))
```

```python
# Find Index of a Target Element
# Problem Statement: Given an array and a target element, return its index.
# Return -1 if element does not exist.

# Example
# Input: arr = [5, 8, 2], target = 8
# Output: 1

def index_finder(arr, target):
    for idx, num in enumerate(arr):
        if num == target:
            return idx
    return -1

print()

```
