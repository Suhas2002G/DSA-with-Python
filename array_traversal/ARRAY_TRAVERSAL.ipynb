```python
# Find Maximum Element in an Array
# Problem Statement: Given an array of integers, find and return the maximum element present in the array.
# Do not use built-in functions like max().
# Input: [3, 5, 1, 9, 2]
# Output: 9

def find_max(arr):
    maximum = arr[0]
    for i in arr:
        if i > maximum:
            maximum = i
    return maximum

print(find_max([3, 5, 1, 9, 2]))
```

```python
# Find Minimum Element in an Array
# Problem Statement: Find the minimum value from a given integer array without using built-in functions.
# Example
# Input: [7, 4, 9, 2, 6]
# Output: 2

def find_min(arr):
    mini = arr[0]
    for i in arr:
        if i < mini:
            mini = i

    return mini

print(find_min([7,4,9,2,6]))
```

```python
# Q3. Sum of All Elements in an Array
# --> Problem Statement: Calculate the sum of all elements present in an integer array.
# --> Example: Input: [1, 2, 3, 4], Output: 10

def find_sum(arr):
    sum = 0
    for i in arr:
        sum += i
    return sum

print(find_sum([1, 2, 3, 4]))
```

```python
# Count Even Numbers in an Array
# Problem Statement: Count how many even numbers are present in the given array.
# Example: Input: [1, 2, 4, 7, 8], Output: 3

def even_counter(arr):
    count = 0
    for i in arr:
        if i % 2 == 0:
            count += 1
    return count
    
arr = [1, 2, 4, 7, 8]
print(even_counter(arr))
```

```python
# Reverse an Array (In-Place)
# Problem Statement: Reverse the array without using extra space.
# Example: Input: [1, 2, 3, 4], Output: [4, 3, 2, 1]
# i used 2-pointer approach

def reverse_array(arr):
    start = 0
    end = len(arr)-1
    while start < end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1
    return arr

print(reverse_array([1, 2, 3, 4, 5]))
```

```python
# Find Index of a Target Element
# Problem Statement: Given an array and a target element, return its index.
# Return -1 if element does not exist.

# Example
# Input: arr = [5, 8, 2], target = 8
# Output: 1

def index_finder(arr, target):
    for idx, num in enumerate(arr):
        if num == target:
            return idx
    return -1

print(index_finder(arr = [1, 5, 8, 2], target = 8))
```

```python
# Check if Array is Sorted (Ascending)
# Problem Statement: Check whether the given array is sorted in non-decreasing order.
# Example
# Input: [1, 2, 3, 4]
# Output: True

def is_sorted(arr):
    curr = 0
    next = 1
    for i in range(len(arr)-1):
        if arr[curr] > arr[next]:
            return False
        next += 1
        curr += 1
    return True

print(is_sorted([1, 2, 3, 4, 23]))
    
```

```python
# Second Largest Element in Array
# Problem Statement: Find the second largest element without sorting.
# Example
# Input: [10, 5, 20, 8]
# Output: 10

def second_largest(arr):
    if len(arr) < 2:
        return None
        
    # Use float('-inf') for a cleaner "negative infinity"
    first = second = float('-inf')
    
    for i in arr:
        if i > first:
            second = first
            first = i 
        elif first > i > second:
            second = i
    return second

print(second_largest([10,10,5,20,8]))

```

```python
# Remove Duplicates from Sorted Array
# Problem Statement: Given a sorted array, remove duplicates and return unique elements.
# Example
# Input: [1, 1, 2, 2, 3]
# Output: [1, 2, 3]

def remove_duplicate(arr):
    result = [arr[0]]
    for i in range(1,len(arr)):
        if arr[i] != arr[i-1]:
            result.append(arr[i])
    return result
    
print(remove_duplicate([1, 1, 2, 2, 3]))

# # NOTE : DONT SOLVE THIS SINCE TIME COMPL. n^2
# def remove_duplicate(arr):
#     uniq = []
#     for num in arr:
#         if num not in uniq:
#             uniq.append(num)

#     return uniq

# print(remove_duplicate([1, 1, 2, 2, 3]))
```

```python
# Move All Zeros to End
# Problem Statement
# Move all 0s to the end without changing order of non-zero elements.
# üîç Example
# [0,1,0,3,12] ‚Üí [1,3,12,0,0]

def move_zeros(arr):
    curr = 0
    for i in range(len(arr)):
        if arr[i] != 0:
            arr[curr],arr[i] = arr[i], arr[curr]
            # print("current arr", arr)
            # print(curr)
            curr+=1
    
    return arr

print(move_zeros([0,1,0,3,12]))
```
